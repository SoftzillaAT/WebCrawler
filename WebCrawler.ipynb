```
# DEFINE CONSTANTS
website = "https://bmtcw.playngonetwork.com/casino/gamehelp?pid=235&gameid=[id]&lang=de_DE&brand=1244&jurisdiction=MT&context=&channel=desktop&version=v2.0.1.179&currency=DEMO"
file_path_out = "/content/drive/My Drive/Colab Notebooks/webcrawler.csv"
min_ID = 1
max_ID = 666
```

```
import urllib.request
import requests
from bs4 import BeautifulSoup
import re
from datetime import date
import os.path
from os import path


def getValuesFromPage(id, url, line):
  page = urllib.request.urlopen(url)
  #print(page.read())

  #html = requests.get(url).text
  #print(html)

  cleantext = BeautifulSoup(page).text

  if ("No help available for the specified game" in cleantext):
    return ""

  # GAME NAME
  lines = cleantext.splitlines()
  game = lines[9]

  # RTP
  index = cleantext.find("Der RTP beträgt")
  if (index < 0):
    rtp = "XXX"
  else:
    rtp = cleantext[index+16:index+21]
  
  #RTP last month
  pattern = "Der RTP für letzten Monat beträgt "
  index = cleantext.find(pattern)
  if (index < 0):
    rtp_last_month = "XXX"
  else:
    rtp_last_month = cleantext[index+len(pattern):index+len(pattern)+5]

  #RTP last 6 month
  pattern = "In den letzten sechs Monaten betrug der RTP dieses Spiels: "
  index = cleantext.find(pattern)
  if (index < 0):
    rtp_last_six_month = "XXX"
  else:
    rtp_last_six_month = cleantext[index+len(pattern):index+len(pattern)+5]
  #print(rtp_last_six_month)
  
  #print(index)
  #print(cleantext)
  print(len(line))
  if len(line) > 0:
    print("append line")
    result = line + ',"' + rtp_last_month.replace(".",",") + '"\r\n'
  else:
    print("new line")
    result = str(id) + ',"' + game + '","' + rtp.replace(".",",") + '","' + rtp_last_month.replace(".",",") + '"\r\n'
  return result
```

```
def getLine(id, lines):
  if len(lines) == 0:
    return ""
  
  for line in lines:
    cells = line.split(",")
    if cells[0] == str(id):
      return line

  return ""
```

```
lines = ""
if path.exists(file_path_out):
  with open(file_path_out) as f:
    lines = f.read().splitlines()

if len(lines) == 0:
  csv = "ID,Game,RTP,Last month @ " + str(date.today()) + "\r\n"
else:
  csv = lines[0] + "," + "Last month @ " + str(date.today()) + "\r\n"

for i in range(min_ID,max_ID):
  link = website.replace("[id]", str(i))
  line = getLine(i, lines)
  csv = csv + getValuesFromPage(i, link, line)
  print("Game " + str(i) + " finished")

f = open(file_path_out, "w")
f.write(csv)
f.close()
```
